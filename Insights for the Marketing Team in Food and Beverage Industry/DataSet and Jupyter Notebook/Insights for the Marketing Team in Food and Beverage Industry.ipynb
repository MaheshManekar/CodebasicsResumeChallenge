# Demographic Insights

```python
# Who prefers energy drink more? (male/female/non-binary?)

gender = merged_Data2.Gender

gender_counts = Counter(gender)

# Pie chart labels and sizes
labels = gender_counts.keys()
sizes = gender_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Gender Distribution')


```

```python
# Which age group prefers energy drinks more?
# Who prefers energy drink more? (male/female/non-binary?)

age = merged_Data2.Age

age_counts = Counter(age)

# Pie chart labels and sizes
labels = age_counts.keys()
sizes = age_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Age Distribution')


```

# Consumer Preferences

```python
# What are the preferred ingredients of energy drinks among respondents?

# Which age group prefers energy drinks more?
# Who prefers energy drink more? (male/female/non-binary?)

Ingredients = merged_Data2.Ingredients_expected

Ingredients_counts = Counter(Ingredients)

# Pie chart labels and sizes
labels = Ingredients_counts.keys()
sizes = Ingredients_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Ingredients Distribution')
```

```python
# What packaging preferences do respondents have for energy drinks?

Packaging = merged_Data2.Packaging_preference

Packaging_counts = Counter(Packaging)

# Pie chart labels and sizes
labels = Packaging_counts.keys()
sizes = Packaging_counts.values()

# Create the pie chart
fig = plt.figure(figsize = (13, 5))
plt.bar(labels, sizes, width=.8)

# Title
plt.title('Packaging Distribution')
plt.show()
```

# Competition Analysis

```python
# Who are the current market leaders?


Brand = merged_Data2.Current_brands

Brand_counts = Counter(Brand)

# Pie chart labels and sizes
labels = Brand_counts.keys()
sizes = Brand_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Brand_perception')
```

# Purchase Behavior

```python
# Where do respondents prefer to purchase energy drinks?

Purchase_Location = merged_Data2.Purchase_location

Purchase_Location_counts = Counter(Purchase_Location)

# Pie chart labels and sizes
labels = Purchase_Location_counts.keys()
sizes = Purchase_Location_counts.values()

# Create the pie chart
fig = plt.figure(figsize = (13, 5))
plt.bar(labels, sizes, width=.8)

# Title
plt.title('Purchase Location')
plt.show()
```

```python
# What are the typical consumption situations for energy drinks among respondents?

Consumption_reason = merged_Data2.Typical_consumption_situations

Consumption_reason_counts = Counter(Consumption_reason)

# Pie chart labels and sizes
labels = Consumption_reason_counts.keys()
sizes = Consumption_reason_counts.values()

# Create the pie chart
fig = plt.figure(figsize = (15, 5))
plt.bar(labels, sizes, width=.8)

# Title
plt.title('Consumption Reason')
plt.show()
```

```python
# What factors influence respondents purchase decisions, such as price range and limited edition packaging?


Price = merged_Data2.Price_range

Price_counts = Counter(Price)

# Pie chart labels and sizes
labels = Price_counts.keys()
sizes = Price_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Price range Purchase')

Limited_Edition = merged_Data2.Limited_edition_packaging

Limited_Edition_counts = Counter(Limited_Edition)

# Pie chart labels and sizes
labels = Limited_Edition_counts.keys()
sizes = Limited_Edition_counts.values()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%.1f%%')

# Title
plt.title('Limited Edition Purchase')
```
